# **If-elif-else-ladder**

```python

mark=35
if (mark>=95) and (mark<=100):
  print('a')
elif (mark>=90) and (mark<=94):
  print('b')
elif (mark>=80) and (mark<=89):
  print('c')
elif (mark>=51) and (mark<=79):
  print('d')
elif (mark>=0) and (mark<=50):
  print('f')
```

# **nested condition**

```python
age=23
citizen=True
if age>=18:
  if citizen:
    print('eligible to vote and citizen')
  else:
    print('eligible to vote but not a citizen')
else:
  print('not eligible to vote and not a citizen')
```
