print values from 1 to 10

````python
i = 1
while i < 21:
  print(i)
  if i==10:
    break
  i += 1
````

# **while with continue**

````python
i = 1
while i < 21:
  if i==10:
    i=i+1
    continue
  print(i)
  i += 1

````
